```python
def extract_surrounding_words(text, keyword, num_words=3):
    words = text.split()
    
    keyword_indices = [i for i, word in enumerate(words) if keyword in word]

    surrounding_words = []
    for index in keyword_indices:
        start = max(0, index - num_words) # Ensure start index isn't negative
        end = min(len(words), index + num_words + 1) # Ensure end index doesn't exceed list bounds
        context = words[start:index] + words[index+1:end] # Exclude the keyword itself
        surrounding_words.append(context)
    
    return surrounding_words

text = "With over four years of industry experience, our team has the knowledge and experience to handle any project."

surrounding_words = extract_surrounding_words(text, "experience")
surrounding_words
```

```python
import re

def extract_years_of_experience(text, keyword="experience"):
    words = text.split()
    
    keyword_indices = [i for i, word in enumerate(words) if keyword in word]
    
    numeric_pattern = re.compile(r'\b\d+\b')
    
    years_of_experience = []
    for index in keyword_indices:
        start = max(0, index - 3) 
        end = min(len(words), index + 4) 
        
        for i in range(start, end):
            if numeric_pattern.match(words[i]):
                years_of_experience.append(words[i])
                break 

    return years_of_experience

text = "With over four years of industry experience, our team has the knowledge and experience to handle any project."

years_of_experience = extract_years_of_experience(text)
years_of_experience
```

```python
import re

def extract_experience_requirements(text):
    # Improved pattern to match a wider range of experience requirements
    # This includes matching:
    # - Direct mentions of years (e.g., "3+ years", "3 years", "three years")
    # - Requirements possibly not directly preceding "years" (e.g., "experience: 3 years")
    pattern = re.compile(r'(\b\d+|\bone|\btwo|\bthree|\bfour|\bfive|\bsix|\bseven|\beight|\bnine|\bten)\s*(\+)?\s*(years)?\s*(of)?\s*(experience)?', re.IGNORECASE)

    # Find all matches in the text
    matches = pattern.findall(text)

    # Process the matches to create a readable list of requirements
    requirements = []
    for match in matches:
        # Formatting the match to a standard form, e.g., "3 years of experience"
        num = match[0]
        plus = "+" if match[1] else ""
        years = " years" if match[2] else " years"
        of_experience = " of experience" if match[4] else " of experience"
        requirement = f"{num}{plus}{years}{of_experience}"
        requirements.append(requirement.strip())

    return list(set(requirements))  # Remove duplicates

text = "Website Developer **About us**We are professional, agile and professional.Our work environment includes:* Modern office setting* Food provided* Work-from-home daysBurt s Wireless Inc., the Southern Tier s premier independently owned and operated MNVO, is currently seeking a talented and dedicated **Website Developer** to join our dynamic team. We re not just another wireless provider; we re a force for positive change in our community. Every month, a portion of our company s profit is used to support local charities. By joining us, you re not just taking a job; you re becoming part of a movement that believes in the incredible power of giving back.Responsibilities:* Collaborate with our team to design, develop, and manage our website and web interfaces.* Develop and optimize both front-end and back-end website elements.* Ensure our website is optimized for mobile and desktop performance.* Conduct regular website testing and debugging.* Maintain a high level of knowledge and expertise in the latest web development industry trends and advancements.Requirements:* Proven experience as a website developer.* Proficient in HTML\/CSS, JavaScript, and other relevant web design coding languages.* Experience with responsive and adaptive design.* Strong problem-solving abilities.* Excellent communication skills and attention to detail.Preferred Qualifications:* Experience with SEO best practices.* Familiarity with content management systems.* Bachelor s degree in computer science or related field.Become a part of our mission to make a difference simply by using your everyday devices. We believe in going beyond the ordinary, and we are looking for candidates who share this passion. If this sounds like you, we would love to hear from you.Remember, at Burt s Wireless Inc., its not just about what you do; its about making a difference in the community we serve. Apply now to become a part of this incredible journey!Job Type: ContractPay: $25.27 - $28.06 per hourExpected hours: 35 per weekBenefits:* Dental insurance* Health insurance* Paid time offSchedule:* Monday to FridaySupplemental pay types:* Bonus opportunities* Commission pay* Signing bonus* TipsExperience:* WordPress: 7 years (Required)Work Location: Remote"
answer = []  # Initialize as an empty list

for text in df_d.head(5):  # Assuming you only want to process the first 5 descriptions
    experience_requirements = extract_experience_requirements(text)

    # In case there are multiple requirements, you might want to join them or handle differently
    requirements_str = "; ".join(experience_requirements) if experience_requirements else "No specific experience requirement found."
    answer.append(requirements_str)

# To display the first 5 answers:
for ans in answer:
    print(ans)

```

```python
def extract_experience_numbers_with_words(text):
    # Mapping of numeric words to their numeric values
    numeric_words = {
        "one": "1", "two": "2", "three": "3", "four": "4", "five": "5",
        "six": "6", "seven": "7", "eight": "8", "nine": "9", "ten": "10",
        "eleven": "11", "twelve": "12", "thirteen": "13", "fourteen": "14",
        "fifteen": "15", "sixteen": "16", "seventeen": "17", "eighteen": "18",
        "nineteen": "19", "twenty": "20"
    }

    # Combine numeric words and digits into a single regex pattern
    number_pattern = re.compile(r'\b(\d+\+?|' + '|'.join(numeric_words.keys()) + r')\b', re.IGNORECASE)
    
    # Tokenize the text into words
    words = text.split()
    
    # Find all occurrences of numeric expressions (digits or numeric words)
    numbers = []
    for i, w in enumerate(words):
        if number_pattern.match(w):
            # Convert numeric words to their corresponding digits
            num = w.lower()
            num = numeric_words.get(num, w)  # Keep the original if it's a digit
            numbers.append((num, i))
    
    # Check for 'experience' or 'years' within 4 words before or after the number
    experience_related_numbers = []
    full = []
    for number, index in numbers:
        # Define the search window
        start = max(0, index - 3)
        end = min(len(words), index + 8)  # +5 because range end is exclusive
        window = words[start:end]
        
        # Check if 'experience' or 'years' is in the window
        if 'experience' in window and any(keyword in window for keyword in ['years', 'year']):
                    full_window = " ".join(window)  # Convert window back to string
                    full.append(full_window)
                    experience_related_numbers.append(number)
            
    return  experience_related_numbers, full
```

```python
def extract_experience_numbers_with_words_completed(text):
    # Mapping of numeric words to their numeric values
    numeric_words = {
        "one": "1", "two": "2", "three": "3", "four": "4", "five": "5",
        "six": "6", "seven": "7", "eight": "8", "nine": "9", "ten": "10",
        "eleven": "11", "twelve": "12", "thirteen": "13", "fourteen": "14",
        "fifteen": "15", "sixteen": "16", "seventeen": "17", "eighteen": "18",
        "nineteen": "19", "twenty": "20"
    }

    # Combine numeric words and digits into a single regex pattern
    number_pattern = re.compile(r'\b(\d+\+?|' + '|'.join(numeric_words.keys()) + r')\b', re.IGNORECASE)
    
    # Tokenize the text into words
    words = text.split()
    
    # Find all occurrences of numeric expressions (digits or numeric words)
    numbers = []
    for i, w in enumerate(words):
        if number_pattern.match(w):
            # Convert numeric words to their corresponding digits
            num = w.lower()
            num = numeric_words.get(num, w)  # Keep the original if it's a digit
            numbers.append((num, i))
    
    # Check for 'experience' or 'years' within 4 words before or after the number
    experience_related_numbers = []
    full = []
    for number, index in numbers:
        # Define the search window
        start = max(0, index - 3)
        end = min(len(words), index + 8)  # +5 because range end is exclusive
        window = words[start:end]
        
        # Check if 'experience' or 'years' is in the window
        if 'experience' in window and any(keyword in window for keyword in ['years', 'year']):
                    full_window = " ".join(window)  # Convert window back to string
                    full.append(full_window)
                    experience_related_numbers.append(number)

    
    highest_value = process_extracted_numbers(experience_related_numbers)
    return  highest_value, experience_related_numbers, full


def process_extracted_numbers(experience_related_numbers):
    processed_numbers = []
    for number in experience_related_numbers:
        # Remove '+' signs
        number = number.replace('+', '')
        # Convert ranges to their lowest value
        if '-' in number:
            number = number.split('-')[0]
        # Convert to integer for comparison
        try:
            processed_number = int(number)
            processed_numbers.append(processed_number)
        except ValueError:
            # Handle the case where the number can't be converted to int
            continue

    # Find the highest value
    if processed_numbers:
        highest_value = max(processed_numbers)
    else:
        highest_value = None

    return highest_value
```
